```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import category_encoders as ce
```

# Q1: Apply Ordinal encoding.

```python
df = pd.DataFrame({'team': ['A', 'A', 'B', 'B', 'B', 'B', 'C', 'C'],
                   'points': [25, 12, 15, 14, 19, 23, 25, 29]})
```

```python
#df = pd.DataFrame({"Score": ["Low", "Low", "Medium", "Medium", "High", "Low", "Medium","High", "Low"]})

dict_Score = {'A':0,'B':1,'C':2}
df['Score_num'] = df['team'].replace(dict_Score)
df
```

# Q2: Apply One hot encoding.

```python

encoder = OneHotEncoder(handle_unknown='ignore')
encoder_df = pd.DataFrame(encoder.fit_transform(df[['team']]).toarray())
final_df = df.join(encoder_df)
final_df
final_df.drop('team',axis=1,inplace=True)
final_df
```

# Q3: Apply Frequency Encoding. 

```python
# Count Encoding - fit on training data, transform test data
encoder1 = ce.CountEncoder(cols="team")
df["team_count_encoded"] = encoder1.fit_transform(df["team"])
df["team_count_encoded"] = encoder1.transform(df["team"])

# Frequency (normalized count) Encoding
encoder2 = ce.CountEncoder(cols="team", normalize=True)
df["team_frequency_encoded"] = encoder2.fit_transform(df["team"])
df["team_frequency_encoded"] = encoder2.transform(df["team"])
df
```

# Q4: Apply Binary Encoding. 

```python
ce_be = ce.BinaryEncoder(cols=['team']);

# transform the data 
data_binary = ce_be.fit_transform(df["team"]);
data_binary
```
